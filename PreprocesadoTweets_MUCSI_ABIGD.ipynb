```python
df.drop(df.columns[[2, 3, 4, 6, 7, 8, 9, 11, 13, 14, 15, 16]], axis=1, inplace=True)
df
```

```python
df = df.drop_duplicates(subset=['text'])
df
```

```python
df['text'] = df['text'].str.lower()
df
```

```python
searchfor = ['chris', 'evans','captain', 'america']
ls = df['text'].str.contains('|'.join(searchfor))
res = [i for i, val in enumerate(ls) if val]
df = df.iloc[res,:]
df
```

```python
df.to_csv('preprocessedTweets.csv')
```
